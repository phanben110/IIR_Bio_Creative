```python
import pandas as pd
import os

def read_score_file(file_path, name_run):
    # Initialize an empty list to store the lines of data
    data_lines = []

    # Open the text file for reading
    with open(file_path, 'r') as file:
        # Flag to indicate when to start storing data
        start_storing = False

        for line in file:
            # Check if the line contains the start marker
            if "==========+Relation type (Details)==========" in line:
                start_storing = True
                continue  # Skip this line

            # If we have encountered the start marker, start storing lines in the list
            if start_storing:
                # Remove leading and trailing whitespace from the line and split it into columns
                columns = line.strip().split('\t')
                # Add the "name_run" column
                columns.append(name_run)
                data_lines.append(columns)

    # Create a pandas DataFrame from the list of data lines
    df = pd.DataFrame(data_lines, columns=["Relation Type", "Presion", "Recall", "F1_Score", "name_run"])

    return df

# File paths for the four score files

# Create DataFrames for each file
```

```python
path = "/Users/benphan/NCKU/IIR-Lab/BioCreative/6_10_final/TestSet/task1_finalPerformance/score_subtask"
file_paths = os.listdir(path)
pair_relation_run = [None, None, None, None]
pair_relation_run[0]=read_score_file(os.path.join(path, file_paths[0]), "run1")
pair_relation_run[1] = read_score_file(os.path.join(path, file_paths[1]), "run2")
pair_relation_run[2] = read_score_file(os.path.join(path, file_paths[2]), "run3")
pair_relation_run[3] = read_score_file(os.path.join(path, file_paths[3]), "run4") 
pair_relation_full = pd.concat(pair_relation_run, ignore_index=True)
```

```python
number_of_relation_type = len(pair_relation_full["Relation Type"].unique().tolist())
print("number_of_relation_type: ", number_of_relation_type)
```

```python
import pandas as pd

# Assuming you have already loaded the pair_relation_full DataFrame

# Convert the "F1_Score" column to a numerical format
pair_relation_full["F1_Score"] = pd.to_numeric(pair_relation_full["F1_Score"])

# Group by "Relation Type" and find the row with the maximum F1_Score within each group
pair_relation_best = pair_relation_full.loc[pair_relation_full.groupby('Relation Type')['F1_Score'].idxmax()]

# Reset the index of the resulting DataFrame
pair_relation_best = pair_relation_best.reset_index(drop=True)
```
